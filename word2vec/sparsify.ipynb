```python
import pandas as pd
import numpy as np
import time 
from scipy import sparse
```

```python
def read_sentence(num_files = 1,sentence_i=1, data_path = ""):
    data = []
    if sentence_i == 1:
        path = data_path + 'Embedded Data/'
        for i in range(num_files):
            filename = 'embedded' + str(i+1) +'.npy'
            temp = np.load(path+filename)
            data.extend(temp)
    else:
        path = data_path + 'Embedded Data 2/'
        for i in range(num_files):
            filename = 'embedded' + str(i+41) +'.npy'
            temp = np.load(path+filename)
            data.extend(temp)
    return data

def sparsify_a_vector(sentence_vec,threshold=0.0005):
    'if value in vector is smaller than threshold, set it to 0'
    for i in range(len(sentence_vec)):
        if sentence_vec[i]<threshold:
            sentence_vec[i] = 0
    #sparse matrix
    return sparse.csr_matrix(sentence_vec)

def sparsify_a_sentence(sentence,threshold=0.0005):
    '''Load in a sentence, return a sparse sentence'''
    sparse_sentence = []
    for vector in sentence:
            sparse_sentence.append(sparsify_a_vector(vector,threshold))
    return sparse_sentence

def sparsify_sentences(sentences,threshold=0.0005):
    sparse_sentences = []
    for sentence in sentences:
        sparse_sentences.append(sparsify_a_sentence(sentence,threshold))
    return sparse_sentences
```

# 存所有file

```python
def load_one_file(file_idx = 1,sentence_i=1, data_path = ""):
    data = []
    if sentence_i == 1:
        path = data_path + 'Embedded Data/'
        filename = 'embedded' + str(file_idx+1) +'.npy'
        temp = np.load(path+filename)
        data.extend(temp)
    else:
        path = data_path + 'Embedded Data 2/'
        filename = 'embedded' + str(file_idx+41) +'.npy'
        temp = np.load(path+filename)
        data.extend(temp)
    return data


def sparsify_all(num_file = 40, sentence_i = 1, path = "E:\ATEC_data/"):
    t = time.time()
    for i in range(num_file):
        sparsied_file = []
        sentences1 = load_one_file(file_idx = i,sentence_i=1, data_path = path)
        sentences2 = load_one_file(file_idx = i,sentence_i=2, data_path = path)
        sparse_s1 = sparsify_sentences(sentences1)
        sparse_s2 = sparsify_sentences(sentences2)
        np.save(path+'sparse1/'+'sparsed'+str(i+1)+'.npy', sparse_s1)
        np.save(path+'sparse2/'+'sparsed'+str(i+41)+'.npy', sparse_s2)
        print('-------done', i, '-----------')
        print('cost time: ',time.time()-t)
```
