```python
def read_sentence(num_files,sentence_i=1):
    data = []
    if sentence_i == 1:
        path = 'Embedded Data/'
        for i in range(num_files):
            filename = 'embedded' + str(i+1) +'.npy'
            temp = np.load(path+filename)
            data.extend(temp)
    else:
        path = 'Embedded Data 2/'
        for i in range(num_files):
            filename = 'embedded' + str(i+41) +'.npy'
            temp = np.load(path+filename)
            data.extend(temp)
    
    return data
```
